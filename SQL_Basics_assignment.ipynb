9. What is the Difference Between a Primary Key and a Foreign Key?

- Key Differences:

Definition & Purpose:
A primary key (e.g., EmployeeID) ensures uniqueness for every row in its table. A foreign key (e.g., DeptID in an Employee table) points to the primary key in a parent table to create a link.
Uniqueness & NULLs: Primary keys cannot be NULL and must be unique. Foreign keys can contain duplicate values and allow NULL values.
Count & Quantity:
A table can have only one primary key constraint, but it can have multiple foreign key constraints.

Referential Integrity: Primary keys prevent duplicate records, while foreign keys prevent actions that would break links between tables (e.g., preventing deletion of a parent record still referenced by a child).

Indexing: Primary keys are automatically indexed for fast retrieval. Foreign keys are not automatically indexed, though it is often recommended.

10. What Are Constraints in SQL and Why Are They Used?

- In SQL, constraints are a set of rules applied to table columns to limit the types of data that can be stored. They act as "gatekeepers," ensuring that every piece of information entered into your database is accurate, reliable, and consistent

Why Constraints Are Used:-

Without these rules, a database could quickly become cluttered with duplicate records, missing essential info, or broken links between related tables. Specifically, they serve to:

Enforce Data Integrity: They guarantee that data remains valid and meaningful throughout its lifecycle.

Prevent Errors: They automatically reject any action (insert, update, or delete) that violates the preset rules.

Maintain Relationships: They ensure that data in one table correctly refers to valid data in another (e.g., an order must belong to an existing customer).

Automate Validation: By handling rules at the database level, they reduce the need for complex manual checks in your application code.
